```python
def clean_merge(df1, df2):
    merged = pd.merge(df1, df2, on='name', how='outer', suffixes=('', '_dup'))

    for col in df1.columns:
        if col != 'name' and col + '_dup' in merged.columns:
            merged[col] = merged[col].combine_first(merged[col + '_dup'])
            merged.drop(columns=[col + '_dup'], inplace=True)

    return merged
```

```python
from sentence_transformers import SentenceTransformer
from sklearn.cluster import KMeans
import numpy as np

# Flatten category lists
all_categories = set(cat.strip() for sublist in final_merged['category_list_cleaned'] for cat in sublist)
all_categories = list(all_categories)

# Step 1: Encode using sentence transformer
model = SentenceTransformer('all-MiniLM-L6-v2')
embeddings = model.encode(all_categories)

# Step 2: Cluster similar categories (try different k)
k = 50  # or adjust based on your need
kmeans = KMeans(n_clusters=k, random_state=42)
clusters = kmeans.fit_predict(embeddings)

# Step 3: Map original categories to cluster labels
category_cluster_map = {cat: f"cluster_{label}" for cat, label in zip(all_categories, clusters)}
```
